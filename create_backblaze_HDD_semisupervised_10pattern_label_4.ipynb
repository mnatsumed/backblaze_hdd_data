```python
import os
import random
import pandas as pd
```

```python
input_data = './all_hdd.csv'
output_base_path = './backblaze_HDD/'
output_path_test = './backblaze_HDD/testing/test_backblazeHDD/'
output_name_rul_end_test = './backblaze_HDD/testing/RUL_test_backblazeHDD.txt'
```

```python
os.makedirs(output_path_test, exist_ok=True)
```

```python
df = pd.read_csv(input_data)
display(df)
```

```python
df = df[['date','serial_number','failure', 'smart_7_norm', 'smart_9_norm', 'smart_240_norm', 'smart_241_norm', 'smart_242_norm']]
```

```python
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].map(lambda x: x.timestamp())
```

```python
random.seed(0)
train_list = random.sample(list(df['serial_number'].unique()), 90)
test_list = list(set(df['serial_number'].unique()) - set(train_list)) #train_list以外の残り件数を抽出
```

```python
test_list
```

```python
for i in range(1,11):
    random.seed(i)
    
    train_list_supervised = random.sample(train_list, 4)
    train_list_unsupervised = list(set(train_list) - set(train_list_supervised))
    
    current_dir = 'training_semi_uniform_1_10_'+str(i)
    output_path_supervised = os.path.join(output_base_path, current_dir, 'supervised')
    output_path_unsupervised = os.path.join(output_base_path, current_dir, 'unsupervised')
    output_name_rul_end = os.path.join(output_base_path, current_dir, 'RUL_end_unsupervised.txt')
    os.makedirs(output_path_supervised, exist_ok=True)
    os.makedirs(output_path_unsupervised, exist_ok=True)
    
    #train_supervised
    for j,name in enumerate(train_list_supervised):
        tmp = df.query('serial_number == @name')[['date','smart_7_norm','smart_9_norm','smart_240_norm','smart_241_norm','smart_242_norm']].sort_values('date')
        tmp.to_csv(os.path.join(output_path_supervised,str(j+1)+'.csv'), index=False)
        
    ## train_unsupervised
    rul_end_supervised = []
    for j,name in enumerate(train_list_unsupervised):
        tmp = df.query('serial_number == @name')[['date','smart_7_norm','smart_9_norm','smart_240_norm','smart_241_norm','smart_242_norm']].sort_values('date')
        random.seed(80*i+j+1)
        del_num = random.randint(1, 10)

        tmp = tmp[:-del_num]
        rul_end_supervised.append(del_num)
        tmp.to_csv(os.path.join(output_path_unsupervised,str(j+1)+'.csv'), index=False)

    with open(output_name_rul_end, mode='w') as f:
        rul_end_supervised = [str(v) for v in rul_end_supervised]
        f.write('\n'.join(rul_end_supervised))
```

```python
rul_end_test = []
for i,name in enumerate(test_list):
    tmp = df.query('serial_number == @name')[['date','smart_7_norm','smart_9_norm','smart_240_norm','smart_241_norm','smart_242_norm']].sort_values('date')
    random.seed(i*100)
    del_num = random.randint(1, 10)
    
    tmp = tmp[:-del_num]
    rul_end_test.append(del_num)
    
    tmp.to_csv(output_path_test+str(i+1)+'.csv', index=False)

with open(output_name_rul_end_test, mode='w') as f:
    rul_end_test = [str(v) for v in rul_end_test]
    f.write('\n'.join(rul_end_test))
```
